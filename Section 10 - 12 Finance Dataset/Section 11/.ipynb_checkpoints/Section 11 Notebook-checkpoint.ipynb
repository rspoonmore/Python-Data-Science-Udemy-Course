```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('dji.csv', parse_dates=['Date'], index_col = 'Date')
```

```python
data.info()
```

```python
df = data.loc['2010-01-01': '2020-03-31', 'Close'].to_frame()
```

```python
df['Return'] = df.pct_change()
```

```python
df.dropna(inplace=True)
```

```python
df.plot(figsize=(20, 12), secondary_y='Return', mark_right=True, fontsize=15)
plt.show()
```

```python
df['Position'] = np.sign(df['Return'])
```

```python
df.head()
```

```python
# Momentum Strategy
df['Strategy_Return'] = df['Position'].shift() * df['Return']
df.head()
```

```python
df['Strategy_Return'].add(1, fill_value=0).cumprod()
df['Strategy'] = df['Strategy_Return'].add(1, fill_value=0).cumprod() * df.iloc[0, 0]
```

```python
df[['Close', 'Strategy']].plot(figsize=(15, 10), fontsize=15)
plt.show()
```

```python
def summary_ann(returns):
    summary = returns.agg(['mean', 'std']).T
    summary['Return'] = summary['mean'] * 252
    summary['Risk'] = summary['std'] * np.sqrt(252)
    summary.drop(columns = ['mean', 'std'], inplace=True)
    return summary
```

```python
summary_ann(df[['Return', 'Strategy_Return']])
```

```python
# Contrarian Strategy
df=data.loc['2010':, 'Close'].to_frame()
df['Return'] = df['Close'].pct_change()
df.dropna(inplace=True)
df.head()
```

```python
df['Position'] = -np.sign(df['Return'])
```

```python
df['Strategy_Return'] = df['Position'].shift() * df['Return']
df['Strategy'] = df['Strategy_Return'].add(1, fill_value=0).cumprod() * df.iloc[0, 0]
```

```python
df[['Close', 'Strategy']].plot(figsize=(15, 10), fontsize=15)
plt.show()
```

```python
summary_ann(df[['Return', 'Strategy_Return']])
```

```python
# Complex strategies
df=data.loc['2010':, 'Close'].to_frame()
df['Return'] = df['Close'].pct_change()
df.dropna(inplace=True)
df.head()
```

```python
df['Position'] = np.where(df['Return'] > 0.0, 0, 1)
df.head()
```

```python
df['Strategy_Return'] = df['Position'].shift() * df['Return']
df['Strategy'] = df['Strategy_Return'].add(1, fill_value=0).cumprod() * df.iloc[0, 0]
df.head()
```

```python
df[['Close', 'Strategy']].plot(figsize=(15, 10), fontsize=15)
plt.show()
```

```python
# Complex strategies
df=data.loc['2010':, 'Close'].to_frame()
df['Return'] = df['Close'].pct_change()
df.dropna(inplace=True)
df['Position'] = np.where(df['Return'] > 0.001, 0, 1)
df['Strategy_Return'] = df['Position'].shift() * df['Return']
df['Strategy'] = df['Strategy_Return'].add(1, fill_value=0).cumprod() * df.iloc[0, 0]
df[['Close', 'Strategy']].plot(figsize=(15, 10), fontsize=15)
plt.show()
```

```python
summary_ann(df[['Return', 'Strategy_Return']])
```

```python
df=data.loc['2010':, 'Close'].to_frame()
df['Return'] = df['Close'].pct_change()
df['SMA50'] = df['Close'].rolling(window=50).mean()
df['SMA200'] = df['Close'].rolling(window=200).mean()
df.dropna(inplace=True)
df['Position'] = np.sign(df['SMA50'] - df['SMA200'])
df['Strategy_Return'] = df['Position'].shift() * df['Return']
df['Strategy'] = df['Strategy_Return'].add(1, fill_value=0).cumprod() * df.iloc[0, 0]
df[['Close', 'Strategy']].plot(figsize=(15, 10), fontsize=15)
plt.show()
```

```python
summary_ann(df[['Return', 'Strategy_Return']])
```
