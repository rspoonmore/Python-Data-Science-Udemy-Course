```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataframes = []
keys = []
for year in range(1880, 2019):
    keys.append(year)
    dataframes.append(pd.read_csv(os.path.join("names",'yob%s.txt' % str(year)), header=None, names=['Name', 'Gender', 'Count']))

df = pd.concat(objs=dataframes, axis=0, keys=keys, names=['Year']).reset_index('Year')
df['Gender'] = df['Gender'].astype('category')
df.head()
```

```python
# Most popular Female names in 2018
df[(df['Year']==2018) & (df['Gender']=='F')].sort_values('Count', ascending=False).head(10)
```

```python
# Evergreen names from 1880 - 2018
originalList = list(df[(df['Year']==1880) & (df['Gender']=='F')].sort_values('Count', ascending=False).head(50)['Name'])
for year in list(df['Year'].unique()):
    updatedList = []
    popularNames = list(df[(df['Year']==year) & (df['Gender']=='F')].sort_values('Count', ascending=False).head(50)['Name'])
    for name in popularNames:
        if name in originalList:
            updatedList.append(name)
    originalList = updatedList
    
print(originalList)
```

```python
totalsDF = df.groupby(['Name', 'Gender'], dropna=True).agg(Total = ('Count', 'sum'),
          No_Year = ('Count', 'count'),
          First_Year = ('Year', 'min'),
          Last_Year = ('Year', 'max'),
          Max_Count = ('Count', 'max')).dropna()
```

```python
# Find best year for each name and gender
bestYearDF = df.copy().set_index(['Name', 'Gender']).merge(totalsDF, left_index=True, right_index=True)
bestYearDF = bestYearDF[bestYearDF['Count'] == bestYearDF['Max_Count']].loc[:, ['Year', 'Count']].drop_duplicates()
```

```python
topNamesByGenderDF = df.groupby("Gender").apply(lambda x: x.groupby('Name')['Count'].sum()).to_frame()
topMaleNames = topNamesByGenderDF[topNamesByGenderDF.index.get_level_values('Gender') == 'M'].sort_values('Count', ascending=False).head(10)
topFemaleNames = topNamesByGenderDF[topNamesByGenderDF.index.get_level_values('Gender') == 'F'].sort_values('Count', ascending=False).head(10)
topNames = pd.concat([topMaleNames, topFemaleNames])
len(topNames)
```

```python
plt.figure(figsize = (15, 10))
sns.barplot(x='Name', y='Count', data=topNames.reset_index(), hue='Gender')
plt.show()
```

```python
babiesPerYear = df.groupby('Year')['Count'].sum()
babiesPerYear.plot(kind='bar', figsize=(15,10))
```

```python
namesPerYear = df.groupby('Year')['Name'].count()
namesPerYear.plot(kind='bar', figsize=(15,10))
```

```python
# Find a name's popularity
namePopDF = df.copy().set_index(['Year', 'Gender'])
namePopDF['Popularity Per Million'] = 1000000 * namePopDF['Count'] / df.groupby(['Year', 'Gender'])['Count'].sum()
df['Rank'] = df.groupby(['Year', 'Gender'])['Count'].apply(lambda x: x.rank(ascending=False)).astype('int')
namePopDF['Rank'] = df.set_index(['Year', 'Gender'])['Rank']
namePopDF.head()
```

```python
maryDF = namePopDF[namePopDF['Name'] == 'Mary'].reset_index('Gender')
maryDF = maryDF[maryDF['Gender'] == 'F'].reset_index()
fig, ax1 = plt.subplots(figsize=(15,10))
ax1.bar(x=maryDF['Year'], height=maryDF['Popularity Per Million'], label='Popularity')
ax2 = ax1.twinx()
ax2.plot(maryDF['Year'], maryDF['Rank'], color='red', label='Rank')
ax1.set_xlabel('Year')
ax1.set_ylabel('Babies per Million')
ax2.set_ylabel('Rank')
ax1.set_title('Mary')
ax1.legend(loc=2)
ax2.legend(loc=1)
plt.show()
```

```python
def plotName(name, gender):
    nameDF = namePopDF[namePopDF['Name'] == name].reset_index('Gender')
    nameDF = nameDF[nameDF['Gender'] == gender].reset_index()
    fig, ax1 = plt.subplots(figsize=(15,10))
    ax1.bar(x=nameDF['Year'], height=nameDF['Popularity Per Million'], label='Popularity')
    ax2 = ax1.twinx()
    ax2.plot(nameDF['Year'], nameDF['Rank'], color='red', label='Rank')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Babies per Million')
    ax2.set_ylabel('Rank')
    ax1.set_title(name)
    ax1.legend(loc=2)
    ax2.legend(loc=1)
    plt.show()
```

```python
plotName('Ryan', 'M')
```

```python
df['Year Total Babies'] = df.merge(df.groupby(['Year', 'Gender'])['Count'].sum(), how='left', left_on=['Year', 'Gender'], right_index=True)['Count_y']
df['Popularity'] = 1000000 * df['Count'] / df['Year Total Babies']
```

```python
# Why does a name's popularity suddenly change?
df['Popularity Lag 1'] = df.groupby(['Name', 'Gender'])['Popularity'].shift()
df.loc[(df['Popularity Lag 1'].isna()) & (df['Year'] > 1880), 'Popularity Lag 1'] = 0
df['Popularity Diff'] = df['Popularity'] - df['Popularity Lag 1']
```

```python
post1900DF = df[df['Year'] >= 1900]
post1900DF.sort_values('Popularity Diff', ascending=False).head(20)
```

```python
plotName('Lisa', 'F')
```

```python
df['%Change'] = 100 * df['Popularity Diff'] / df['Popularity Lag 1']
df.loc[(df['Popularity Lag 1'] == 0), '%Change'] = np.nan
df.sort_values('%Change', ascending=False).head(10)
```

```python
# Find the most consistent names
agg = totalsDF.reset_index()
agg['Spike Score'] = 100 * agg['Max_Count'] / agg['Total']
agg.sort_values('Spike Score', ascending=True).head(10)
```

```python
# Most common unisex names of all time
femaleNameTotalDF = df[df['Gender'] == 'F'].groupby('Name')['Count'].sum().to_frame()
nameTotalDF = df.groupby('Name')['Count'].sum().to_frame()
nameCompDF = femaleNameTotalDF.merge(nameTotalDF, how='inner', left_index=True, right_index=True, suffixes=('_Female', '_Total'))
nameCompDF['Count_Male'] = nameCompDF['Count_Total'] - nameCompDF['Count_Female']
nameCompDF = nameCompDF[nameCompDF['Count_Male'] > 0]
nameCompDF['Name MF Gender Ratio'] = nameCompDF['Count_Male'] / nameCompDF['Count_Female']
unisexNameDF = nameCompDF[(nameCompDF['Name MF Gender Ratio'] < 1) & (nameCompDF['Name MF Gender Ratio'] > 0.9)]
mask1 = df.apply(lambda row: row['Name'] in list(unisexNameDF.index), axis=1)
df[mask1].groupby(['Name', 'Gender'])['Count'].sum()
```
