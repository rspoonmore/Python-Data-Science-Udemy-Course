```python
import pandas as pd
import os
from glob import glob
```

```python
df = pd.read_csv(os.path.join("names","yob1880.txt"), header=None, names=['Name', 'Gender', 'Count'])
df.head()
```

```python
dataframes = []
keys = []
for year in range(1880, 2019):
    keys.append(year)
    dataframes.append(pd.read_csv(os.path.join("names",'yob%s.txt' % str(year)), header=None, names=['Name', 'Gender', 'Count']))

df = pd.concat(objs=dataframes, axis=0, keys=keys, names=['Year']).reset_index('Year')
df.head()
```

```python
filenames = glob(os.path.join('namesbystate','*.TXT'))
#for file in glob(os.path.join('namesbystate','subdir','*.txt')):
#    filenames.append(file)
```

```python
dataframes = []
for file in filenames:
    newDF = pd.read_csv(file, header=None, names=['State', 'Gender', 'Year', 'Name', 'Count'])
    dataframes.append(newDF)
    
namesByStateDF = pd.concat(dataframes, ignore_index=True)
```

```python
namesByStateDF.head()
```

```python
# How to save memory
namesByStateDF.nunique()
print(namesByStateDF.info())
namesByStateDF['Gender'] = namesByStateDF['Gender'].astype('category')
namesByStateDF['State'] = namesByStateDF['State'].astype('category')
print(namesByStateDF.info())
```
